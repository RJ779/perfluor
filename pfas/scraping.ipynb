```python
import requests
from bs4 import BeautifulSoup
import re

url = "https://nwis.waterdata.usgs.gov/nwis/qwdata?state_cd=ma&obs_date_range=1&qw_begin_date=08/01/2020&qw_end_date=2020-12-15&multiple_parameter_cds=53961%2C53962%2C54113%2C54116%2C54117%2C54220%2C54221%2C54222%2C54223%2C54224%2C54225%2C54226%2C54227%2C54228%2C54229%2C54230%2C54231%2C54232%2C54233%2C54234%2C54235%2C54236%2C54237%2C54238&param_cd_operator=AND&medium_cd=WS&group_key=NONE&sitefile_output_format=html_table&column_name=station_nm&inventory_output=0&rdb_inventory_output=file&TZoutput=0&pm_cd_compare=Greater%20than&radio_parm_cds=previous_parm_cds&format=html_table&qw_attributes=0&qw_sample_wide=wide&rdb_qw_attributes=0&date_format=YYYY-MM-DD&rdb_compression=file&list_of_search_criteria=state_cd%2Cobs_date_range%2Cmedium_cd%2Cmultiple_parameter_cds"

response = requests.get(url)
response.raise_for_status()

soup = BeautifulSoup(response.content, 'html.parser')

# Initialize empty lists to hold the data
names = []
pfhxs = []
pfoa = []
pfos = []
pfbs = []
pfba = [] 
pfds = []
pfda = []
pfdoa = []
pfhps = [] 
pfhpa = [] 
pfhxa = [] 
pfns = []
pfna = [] 
pfosa = []
pfpes = [] 
pfpea = []
pfta = [] 
pftrda = []
pfunda = [] 

# Extract water body names
water_bodies = soup.find_all('h3')
for water_body in water_bodies:
    name = water_body.get_text(strip=True)
    names.append(name)

# Extract coordinates
latitudes = []
longitudes = []

coordinates_divs = soup.find_all('div', align='left')
for coordinates_div in coordinates_divs:
    text = coordinates_div.get_text(strip=True).lower()
    lat_match = re.search(r'latitude\s*([0-9]+°[0-9]+\'[0-9.]+")', text)
    lon_match = re.search(r'longitude\s*([0-9]+°[0-9]+\'[0-9.]+")', text)
    if lat_match and lon_match:
        lat = lat_match.group(1)
        lon = lon_match.group(1)
        latitudes.append(lat)
        longitudes.append(lon)
```

```python
# Print the lists to verify
print("Latitudes:", latitudes)
print("Longitudes:", longitudes)
```

```python
from datetime import datetime

td_elements = soup.find_all('td')

def is_date_td(td):
    try:
        date_time_str = td.get_text(strip=True)
        datetime.strptime(date_time_str, '%Y-%m-%d %H:%M')
        return True
    except ValueError:
        return False

# List to store the indices of date elements
date_indices = []

# Find the indices of all date elements
for i, td in enumerate(td_elements):
    if is_date_td(td):
        date_indices.append(i)

for index in range(0, len(date_indices), 3): 
    start_index = date_indices[index]
    print(f'Starting at date: {td_elements[start_index].get_text(strip=True)}')
    if start_index + 7 < len(td_elements):
        hxs = td_elements[start_index + 7].get_text(strip=True)
        print(f'pfhxs: {hxs}')
        pfhxs.append(hxs)
    if start_index + 8 < len(td_elements): 
        oa = td_elements[start_index + 8].get_text(strip=True)
        print(f'pfoa: {oa}')
        pfoa.append(oa)
    if start_index + 9 < len(td_elements):
        os = td_elements[start_index + 9].get_text(strip=True)
        print(f'pfos: {os}')
        pfos.append(os)
    if start_index + 13 < len(td_elements): 
        bs = td_elements[start_index + 13].get_text(strip=True)
        print(f'pfbs: {bs}')
        pfbs.append(bs)
    if start_index + 14 < len(td_elements): 
        ba = td_elements[start_index + 14].get_text(strip=True)
        print(f'pfba: {ba}')
        pfba.append(ba)
    if start_index + 15 < len(td_elements): 
        ds = td_elements[start_index + 15].get_text(strip=True)
        print(f'pfds: {ds}')
        pfds.append(ds)
    if start_index + 16 < len(td_elements): 
        da = td_elements[start_index + 16].get_text(strip=True)
        print(f'pfda: {da}')
        pfda.append(da)
    if start_index + 17 < len(td_elements): 
        doa = td_elements[start_index + 17].get_text(strip=True)
        print(f'pfdoa: {doa}')
        pfdoa.append(doa)
    if start_index + 18 < len(td_elements): 
        hps = td_elements[start_index + 18].get_text(strip=True)
        print(f'pfhps: {hps}')
        pfhps.append(hps)
    if start_index + 19 < len(td_elements): 
        hpa = td_elements[start_index + 19].get_text(strip=True)
        print(f'pfhpa: {hpa}')
        pfhpa.append(hpa)
    if start_index + 20 < len(td_elements): 
        hxa = td_elements[start_index + 20].get_text(strip=True)
        print(f'pfhxa: {hxa}')
        pfhxa.append(hxa)
    if start_index + 21 < len(td_elements): 
        ns = td_elements[start_index + 21].get_text(strip=True)
        print(f'pfns: {ns}')
        pfns.append(ns)
    if start_index + 22 < len(td_elements): 
        na = td_elements[start_index + 22].get_text(strip=True)
        print(f'pfna: {na}')
        pfna.append(na)
    if start_index + 23 < len(td_elements): 
        osa = td_elements[start_index + 23].get_text(strip=True)
        print(f'pfosa: {osa}')
        pfosa.append(osa)
    if start_index + 24 < len(td_elements): 
        pes = td_elements[start_index + 24].get_text(strip=True)
        print(f'pfpes: {pes}')
        pfpes.append(pes)
    if start_index + 25 < len(td_elements): 
        pea = td_elements[start_index + 25].get_text(strip=True)
        print(f'pfpea: {pea}')
        pfpea.append(pea)
    if start_index + 26 < len(td_elements): 
        ta = td_elements[start_index + 26].get_text(strip=True)
        print(f'pfta: {ta}')
        pfta.append(ta)
    if start_index + 27 < len(td_elements): 
        trda = td_elements[start_index + 27].get_text(strip=True)
        print(f'pftrda: {trda}')
        pftrda.append(trda)
    if start_index + 28 < len(td_elements): 
        unda = td_elements[start_index + 28].get_text(strip=True)
        print(f'pfunda: {unda}')
        pfunda.append(unda)
```
